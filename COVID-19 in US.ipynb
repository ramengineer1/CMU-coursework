```python
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
import pandas as pd
%matplotlib inline
```

```python
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 50)
```

```python
#database is updated/maintained by Johns Hopkins University Coronavirus Resource Center 
#https://coronavirus.jhu.edu/map.html
#https://github.com/CSSEGISandData/COVID-19
#https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv

url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv"
urllib.request.urlretrieve(url,"time_series_covid19_confirmed_US.csv")
df = pd.read_csv("time_series_covid19_confirmed_US.csv")
#df
```

## U.S.

```python
US_confirmed_cases = np.array(df.sum()[10:])
rate_US_confirmed_cases = np.gradient(US_confirmed_cases, edge_order = 2)
day = len(US_confirmed_cases) #days since first infection
day_span = np.arange(0,day)

#1st infection occurred on 1/22/2020 in US
```

```python
#confirmed cases
plt.plot(day_span,US_confirmed_cases,'o')
plt.title("Confirmed Case in US")
plt.xlabel("Days since 1st Infection")
plt.ylabel("Confirmed Cases")
plt.xlim([50,day_span[-1]+1])
plt.grid()
plt.show()
```

```python
#rate of increase of confirmed cases
plt.plot(day_span,rate_US_confirmed_cases,'o')
plt.title("Confirmed Cases Rate of Increase in US")
plt.xlabel("Days since 1st Infection")
plt.ylabel("Rate of Increase (Case/day)")
plt.xlim([50,day_span[-1]+1])
plt.grid()
plt.show()
```
